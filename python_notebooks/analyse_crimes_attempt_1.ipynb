```python
#Import important libraries for the analysis

import plotly
from plotly.graph_objs import Scatter, Heatmap, Layout
plotly.offline.init_notebook_mode(connected=True)

import numpy as np
import pandas as pd
import scipy
import collections
import datetime
```

```python
#Reading the file.

file = pd.read_csv("improved_spvm_2015-2017.csv")
```

```python
# Crimes per day

jours = file["JOUR"]
jours = list(jours)
jours = collections.Counter(jours)

jours_x = list(jours.keys())
jours_y = list(jours.values())

plotly.offline.iplot({
    "data": [Scatter(x= jours_x, y= jours_y)],
    "layout": Layout(title="Crimes per day")
})
```

```python
#Generating maps for analysis.

d = {"ZONE":file["ARRONDISSEMENT"], "LAT": file["LAT"], "LONG": file["LONG"]}
df = pd.DataFrame(data=d)

df.to_csv("dpt_nbr2.csv", sep=',')
```

```python
#Crimes per day and day division

dates = file["JOUR"]
jours = [ datetime.datetime.strptime(dates[x], '%Y-%m-%d').strftime('%A') for x in range(0, len(dates))]

d = {"JOUR":jours,"QUART":file["QUART"]}
df = pd.DataFrame(data=d)

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday','Sunday']
donnees = []

for i in range(0, len(days)):
    jour = len(df[(df['JOUR'] == days[i]) & (df['QUART'] == 'jour')])
    nuit = len(df[(df['JOUR'] == days[i]) & (df['QUART'] == 'nuit')])
    soir = len(df[(df['JOUR'] == days[i]) & (df['QUART'] == 'soir')])
    data = [jour,soir,nuit]
    donnees.append( data )

print(donnees)
    
plotly.offline.iplot({
    "data": [Heatmap(z=donnees, y=days, x=['8.01am - 4pm', '4.01pm - 12am', '12.01am - 8.00am'], colorscale='Viridis')],
    "layout": Layout(title="Crimes per day-slot")
})
```
